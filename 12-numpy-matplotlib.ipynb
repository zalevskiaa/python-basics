## Микровведение в массивы

Можно просто взять список чисел и создать из него массив

```python
listik = [1, 2, 3, 4, 5, 25]
a = np.array(listik)

print(a)
```

Смотрите, у массива к числам можно обращаться по индексам, так же, как и у списка!

```python
print(a[0])
print(a[-1])
print(a[2:4])
```

Короче, давайте **аккуратно** скажем, что массив это типа такой список чисел.

А теперь давайте посмотрим, чего не умеют списки, но умеют массивы

```python
print(a + 100)
print(a * 100)
print(a * a)
print(a ** 2)
```

```python
b = a ** 2

print(a)
print(b)
print(a + b)
```

Короче, очень круто. Применяем математическую операцию к массиву, а она применяется к каждому числу в массиве.  

А для списков пришлось бы писать компрэхэншены:  
`spisok_2 = [e ** 2 for e in spisok]`

Вот ещё несколько крутых фишек массивов, которыми мы будем пользоваться

- `np.linspace(start, stop, num=50)` - создаёт массив из `num` чисел от `start` до `end` с равным шагом

```python
print(np.linspace(0, 10, 11))
print(np.linspace(0, 4, 21))
```

- `array.shape` - "форма" массива - кортеж из всех размеров массива.

Так как у нас одномерный массив, кортеж содержит одно число - размер этого массива

```python
print(a.shape)
print(type(a.shape))
```

- `zeros(shape)` - Принимает аргумент `shape` - кортеж из чисел - и создаёт массив формы `shape`.  

```python
print(np.zeros(a.shape))
print(np.zeros((3, 4)))
```

## 2D plots

### plot

Берём массив иксов и массив игреков и вызвваем `ax.plot`

```python
x = np.linspace(-5, 5, 44)          # Создаётся массив из 44 чисел от -5 до 5
y = x ** 4 - 20 * x ** 2 + 50       # Для каждого числа вычисляется такая функция


# Фигакс!
# fig - figure - рисунок, ax - axes - оси
# На рисунке находятся оси, на осях рисуется график
fig, ax = plt.subplots()

# на осях рисуется график
ax.plot(x, y)

# В конце пишется вот такая штука.
# show - показать
plt.show()
```

### scatter

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

### stem

```python
fig, ax = plt.subplots()
ax.stem(x, y)
plt.show()
```

### fill_between

В отличие от предыдущих примеров, берём массив иксов и два массива игреков

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x) * 100

fig, ax = plt.subplots()
ax.plot(x, y, color='black')
ax.fill_between(x, y - 20, y + 20, alpha=0.3, color='red')
ax.fill_between(x, y - 70, y + 70, alpha=0.2, color='red')
plt.show()
```

### bar

Перечисляем иксы (не обязательно числа) и игреки, и вызываем `ax.bar`

```python
fig, ax = plt.subplots()
ax.bar(x, y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y, width=0.8 * (x[1] - x[0]))
plt.show()
```

### Несколько графиков на одних осях

дописываем рядом ещё один `ax.plot`

```python
x = np.linspace(-5, 5, 44)
y = x ** 4 - 20 * x ** 2 + 100
y2 = 4 * x ** 3 - 40 * x

fig, ax = plt.subplots()

ax.plot(x, y)
ax.plot(x, y2)

plt.show()
```

### design

```python
t = np.linspace(0, 2 * np.pi, 30)
x = np.sqrt(2) * np.sin(t) ** 3
y = -np.cos(t) ** 3 - np.cos(t) ** 2 + 2 * np.cos(t) 

fig, ax = plt.subplots()

# color - цвет - может задаваться именем, а может тремя числами от 0 до 1 в формате RGB
# label - метка - как будет подписываться график на легенде
# marker - маркер - каким значком будет отображаться каждая точка на графике
# linestyle - стиль линии - каким стилем будет рисоваться линия
# linewidth - толщина линии

ax.plot(x, y, color=(0.2, 0.4, 1), label='это первый график', marker='s', linestyle='dotted', linewidth=3)
ax.plot(x + 3, y, color='red', label='это второй график', marker='o')

ax.set_xlim(-2, 5)              # границы графика по х
ax.set_ylim(-3, 2)              # границы графика по у
ax.set_title('Это заголовок')   # заголовок
ax.set_xlabel('это ось x')      # подпись оси х
ax.set_ylabel('это ось y')      # подпись оси у
ax.set_facecolor("mistyrose")   # цвет фона
ax.legend()                     # отображение легенды
ax.grid()                       # отображение сетки

plt.show()
```

### serdechko

```python
def set_ax_same_lims(ax):
    lim = max(map(abs, ax.get_xlim() + ax.get_ylim() + ax.get_zlim()))

    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)
    ax.set_zlim(-lim, lim)
```

```python
u = np.linspace(0, 2 * np.pi, 30)
v = np.linspace(0, np.pi, 10)
U, V = np.meshgrid(u, v)

X = np.sin(V) * (15 * np.sin(U) - 4 * np.sin(3 * U))
Y = 8 * np.cos(V)
Z = np.sin(V) * (15 * np.cos(U) - 5 * np.cos(2 * U) - 2 * np.cos(3 * U) - np.cos(4 * U))           


fig, ax = plt.subplots(subplot_kw=dict(projection="3d"))
plot_geeks = ax.plot_surface(X, Y, Z, cmap='magma')

set_ax_same_lims(ax)    # делаем границы графика одинаковыми, чтобы сохранить пропорции
plt.show() 
```

## Sliders

```python
a, b, c = 1, 0, 0
x = np.linspace(-10, 10, 50)
y = a * (x - b) ** 2 + c


fig, ax = plt.subplots()


line, = ax.plot(x, y, linewidth=2, color=(0.8, 0, 0))

fig.subplots_adjust(bottom=0.2)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.grid()

a_slider = Slider(fig.add_axes([0.2, 0.10, 0.6, 0.03]), 'a', -2, 2, valinit=a, valstep=0.05)
b_slider = Slider(fig.add_axes([0.2, 0.05, 0.6, 0.03]), 'b', -10, 10, valinit=b, valstep=0.05)
c_slider = Slider(fig.add_axes([0.2, 0.00, 0.6, 0.03]), 'c', -10, 10, valinit=c, valstep=0.05)

def update(val):
    a, b, c = a_slider.val, b_slider.val, c_slider.val
    
    y = a * (x - b) ** 2 + c
    line.set_ydata(y)

    fig.canvas.draw_idle()


a_slider.on_changed(update)
b_slider.on_changed(update)
c_slider.on_changed(update)
plt.show()
```

## Несколько графиков на одном рисунке

`fig, ax = plt.subplots()` - один рисунок и одни оси  
`fig, ax = plt.subplots(nrows, ncols)` - один рисунок и `nrows * ncols` осей


```python
nrows, ncols = 1, 2
a1, a2 = 3, 8

x = np.linspace(-np.sqrt(3), np.sqrt(3), 100)
y = np.cbrt(x ** 2) + np.sqrt(3.3 - x ** 2) * np.sin(a1 * np.pi * x)
y2 = np.cbrt(x ** 2) + np.sqrt(3.3 - x ** 2) * np.sin(a2 * np.pi * x)

fig, ax = plt.subplots(nrows, ncols)
fig.set_size_inches(8, 3)

ax[0].plot(x, y)
ax[1].plot(x, y2)

plt.show()
```

```python
nrows, ncols = 4, 4
a_values = np.linspace(0.001, 8, nrows * ncols)

fig, ax = plt.subplots(nrows, ncols)
fig.set_size_inches(8, 6)
fig.tight_layout()

for i in range(nrows):
    for j in range(ncols):
        a = a_values[ncols * i + j]

        x = np.linspace(-np.sqrt(3), np.sqrt(3), 100)
        y = np.cbrt(x ** 2) + np.sqrt(3.3 - x ** 2) * np.sin(a * np.pi * x)

        ax[i, j].plot(x, y)

        ax[i, j].set_title(f'a = {a :.3f}')
        ax[i, j].set_yticklabels([])
        ax[i, j].set_xticklabels([])
        ax[i, j].margins(0.2)

plt.show()
```
